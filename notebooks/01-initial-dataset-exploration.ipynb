# Solve empty lists not as nan

```python
movie = df.iloc[1].movies
print(movie)
print(type(movie))
```

List are converted to str objects in pandas, so empty lists are just strings of two values. All fields like these are going to need cleaning
(movies, meta_keywords, kb_keywords)

```python
import numpy as np

def str_to_list(df, column_name):
    df = df.copy()
    df[column_name] = df[column_name].apply(eval)
    df[column_name] = df[column_name].apply(lambda x: np.nan if len(x) == 0 else x)
    return df
```

```python
# Actual listings with videos
list_columns = ['movies', 'meta_keywords', 'kb_keywords']
for column in list_columns:
    df = str_to_list(df, column)
df.isna().sum()
```
